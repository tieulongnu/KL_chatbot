```python
from underthesea import word_tokenize
with open("cauhoi_tok.txt","w") as of:
    with open("cauhoi.txt","r") as rf:
        lines = rf.readlines()
        for line in lines:
            line = word_tokenize(line,format="text")
            print(line)
            of.write(line+"\n")
```

```python
keywords= set()
with open("keywords.txt","r",encoding="utf8") as of:
    with open("cauhoi_tok.txt","r",encoding="utf8") as rf:
        lines = rf.readlines()
        for line in lines:
            words = line.strip().split()
            for i in words:
                keywords.add(i)
        for w in sorted(keywords):            
            print(w.lower())          
```

```python
keywordstoint = {}
inttokeywords = {}
keywords=[]
with open("keywords.txt","r",encoding="utf8") as rf:
    lines = rf.readlines()
    for line in lines:
        keywords.append(line.strip())
        
for i,line in enumerate(keywords):
    keywordstoint[line] = i
    inttokeywords[i] = line
        
```

```python
len(keywords)
```

```python
import numpy as np
def TF_IDF(sentence):
    words = sentence.strip().lower().split()
    embedding = np.zeros(1016,dtype = "int32")
    for word in words:
        if word in keywords:
            index = keywordstoint[word]
            embedding[index] +=1 
    return embedding
```

```python
k = TF_IDF("Trường Đại_học Sư_phạm Ngoại_ngữ Hà_Nội được phát_triển trên cơ_sở Khoa nào của trường Đại_học Sư_phạm Hà_Nội ")
```

```python
np.linalg.norm(k)
```
